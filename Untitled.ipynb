```python
import pandas as pd
import matplotlib.pyplot as plt

# Option 1 : Lire seulement les 1000 premières lignes pour tester
df = pd.read_csv("survey_results_public.csv")
print("Sample loaded:")
print(df.head())


```

```python
print("Nombre de lignes :", df.count())
```

```python
# Sélection des colonnes spécifiques dans le DataFrame
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]

# Renommage de la colonne "ConvertedCompYearly" en "Salary" pour plus de clarté
df = df.rename(columns={"ConvertedCompYearly": "Salary"})

# Affichage des 5 premières lignes du DataFrame
print("DataFrame après sélection et renommage :")
display(df.head())

# Filtrage pour ne garder que les lignes où le
df=df[df["Salary"].notnull()]
display(df.head())
df=df.dropna()
df.isnull().sum()
# Filtrer pour ne garder que les employés à temps plein
df = df[df["Employment"] == "Employed, full-time"]

# Supprimer la colonne Employment (avec inplace=True pour modifier directement le DataFrame)
df.drop("Employment", axis=1, inplace=True)

# Vérifier les informations du DataFrame
df.info()
df["Country"].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map  # ⚠️ Le return doit être **en dehors** de la boucle !

# Appliquer la fonction
country_map = shorten_categories(df['Country'].value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
print(df['Country'].value_counts())
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,7))
df.boxplot('Salary','Country',ax=ax)
plt.suptitle('Salary (USS) V Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df=df[df["Salary"]<=250000]
df=df[df["Salary"]>=10000]
df=df[df["Country"]!='Other']
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,7))
df.boxplot('Salary','Country',ax=ax)
plt.suptitle('Salary (USS) V Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
def clean_experience(x):
    if x=='More than 50 years':
        return 50
    if x=='Less than 1 year':
        return 0.5
    return float(x)
df['YearsCodePro']=df['YearsCodePro'].apply(clean_experience)
         
    
```

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
    x = str(x).lower()  # Convertit en minuscules et gère les NaN
    if "bachelor" in x or "b.a." in x or "b.s." in x or "b.eng" in x:
        return "Bachelor's degree"
    if "master" in x or "m.a." in x or "m.s." in x or "mba" in x or "m.eng" in x:
        return "Master's degree"
    if "professional" in x or "doctoral" in x or "ph.d" in x or "ed.d" in x or "jd" in x or "md" in x:
        return "Post grad"
    return "Less than a Bachelors"  # Tout le reste
```

```python
df["EdLevel"] = df["EdLevel"].apply(clean_education)

df["EdLevel"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education=LabelEncoder()
df["EdLevel"]=le_education.fit_transform(df["EdLevel"])
df["EdLevel"].unique()
```

```python
le_country=LabelEncoder()
df["Country"]=le_country.fit_transform(df["Country"])
df["Country"].unique()
```

```python
x=df.drop("Salary",axis=1)
y=df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression 
linear_reg= LinearRegression ()
linear_reg.fit(x,y.values)
y_pred=linear_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error ,mean_absolute_error
import numpy as np
error=np.sqrt(mean_squared_error(y,y_pred))
print (error)
```

```python
from sklearn.tree import DecisionTreeRegressor 
dec_tree_reg=DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(x,y.values)
DecisionTreeRegressor(random_state=0)
```

```python
y_pred=dec_tree_reg.predict(x)
error=np.sqrt(mean_squared_error(y,y_pred))
print(error)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_fore_regr=RandomForestRegressor(random_state=0)
random_fore_regr.fit(x,y.values)
y_pred=random_fore_regr.predict(x)
error=np.sqrt(mean_squared_error(y,y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV
max_depth=[None,2,4,6,8,10,12]
parameters={"max_depth":max_depth}
regressor=DecisionTreeRegressor(random_state=0)
gs=GridSearchCV(regressor,parameters,scoring='neg_mean_squared_error')
gs.fit(x,y.values)
```

```python
regressor=gs.best_estimator_
regressor.fit(x,y.values)
y_pred=regressor.predict(x)
error=np.sqrt(mean_squared_error(y,y_pred))
print("${:,.02f}".format(error))
```
